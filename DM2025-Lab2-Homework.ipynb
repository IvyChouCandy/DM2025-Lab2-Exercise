## 1. Preprocessing Steps

```python
import json
import pandas as pd

def process_emotion_data():
    """csvtrain&test"""

    # 
    with open('./data_phase3/final_posts.json', 'r', encoding='utf-8') as f:
        raw_data = json.load(f)

    # 
    parsed_data = []
    for entry in raw_data:
            post_data = entry['root']['_source']['post']
            parsed_data.append({
                'id': post_data['post_id'],
                'text': post_data['text']
            })

    #  DataFrame
    df_tweets = pd.DataFrame(parsed_data)
    print(f"JSON  {len(df_tweets)} ")

    # data
    df_split = pd.read_csv('./data_phase3/data_identification.csv') 
    df_labels = pd.read_csv('./data_phase3/emotion.csv')       
    df_submit_req = pd.read_csv('./data_phase3/samplesubmission.csv') 
    
    # 
    df_full = df_tweets.merge(df_labels, on='id', how='left')
    
    #  (train/test split)
    df_full = df_full.merge(df_split, on='id', how='left')
    
    df_train = df_full[df_full['split'] == 'train'].copy()
    df_train = df_train[['id', 'text', 'emotion']]
    df_train.to_csv('./data_phase3/train_processed.csv', index=False, encoding='utf-8-sig')

    df_test = df_full[df_full['split'] == 'test'].copy()
    df_test = df_test[['id', 'text', 'emotion']]
    df_test.to_csv('./data_phase3/test_processed.csv', index=False, encoding='utf-8-sig')

    df_submit = df_submit_req[['id']].merge(df_tweets, on='id', how='left')
    
    #  ID  text
    if df_submit['text'].isnull().any():
        print(" ID  JSON ")
    
    df_submit = df_submit[['id', 'text']]
    df_submit.to_csv('./data_phase3/submission_processed.csv', index=False, encoding='utf-8-sig')

if __name__ == "__main__":
    process_emotion_data()
```

#### Clean data

```python
import pandas as pd
import re
import emoji


def clean_tweet_text(text):
    if not isinstance(text, str):
        return ""
    
    # HTML
    text = text.replace("&amp;", "and").replace("&lt;", "<").replace("&gt;", ">")
    
    # 
    text = text.lower()
    
    # 
    #    ":face_with_tears_of_joy:"
    text = emoji.demojize(text, delimiters=(" ", " "))
    
    # 
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # 
    text = re.sub(r'@\w+', '', text)
    
    #  Hashtags -  #
    text = re.sub(r'#(\w+)', r'\1', text)
    
    # 
    contractions = {
        "i'm": "i am", "he's": "he is", "she's": "she is", "it's": "it is",
        "that's": "that is", "what's": "what is", "where's": "where is",
        "there's": "there is", "who's": "who is", "how's": "how is",
        "\'ll": " will", "\'ve": " have", "\'re": " are", "\'d": " would",
        "n't": " not", "won't": "will not", "can't": "cannot"
    }
    for word, replacement in contractions.items():
        text = text.replace(word, replacement)
    
    #  (Ex: sooooo -> so)
    #  3  2 
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    # 
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

def run_cleaning_pipeline():
    df_train = pd.read_csv('./data_phase3/train_processed.csv')
    df_test = pd.read_csv('./data_phase3/test_processed.csv')
    df_submit = pd.read_csv('./data_phase3/submission_processed.csv')

    # clean_tweet_text
    df_train['clean_text'] = df_train['text'].apply(clean_tweet_text)
    df_test['clean_text'] = df_test['text'].apply(clean_tweet_text)
    df_submit['clean_text'] = df_submit['text'].apply(clean_tweet_text)
    df_train = df_train[['id', 'clean_text', 'emotion']].rename(columns={'clean_text': 'text'})
    #     
    df_train.to_csv('./data_phase3/train_cleaned.csv', index=False, encoding='utf-8-sig')
    df_test.to_csv('./data_phase3/test_cleaned.csv', index=False, encoding='utf-8-sig')
    df_submit.to_csv('./data_phase3/submission_cleaned.csv', index=False, encoding='utf-8-sig')
    

if __name__ == "__main__":
    run_cleaning_pipeline()
```

#### Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
from collections import Counter

def visualize_data(file_path):

    df = pd.read_csv(file_path)

    # text
    df['text'] = df['text'].astype(str)

    # 
    sns.set(style="whitegrid")
    
    # 
    plt.figure(figsize=(10, 6))
    # 
    sns.countplot(data=df, x='emotion', order=df['emotion'].value_counts().index, palette='viridis')
    plt.title('Emotion Label Distribution', fontsize=15)
    plt.xlabel('Emotion', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # 
    df['word_count'] = df['text'].apply(lambda x: len(x.split()))
    
    plt.figure(figsize=(12, 6))
    sns.histplot(df['word_count'], bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Tweet Lengths (Word Count)', fontsize=15)
    plt.xlabel('Number of Words', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # 
    print("")
    print(df['word_count'].describe())
    
    # 
    all_text = " ".join(sentence for sentence in df['text'])
    
    # 
    # 
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(width=800, height=400, 
                          background_color='white', 
                          stopwords=stopwords,
                          min_font_size=10).generate(all_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title('Most Common Words (Word Cloud)', fontsize=15)
    plt.show()

    # 20
    # 
    words = all_text.split()
    #  
    filtered_words = [word for word in words if word not in stopwords]
    word_counts = Counter(filtered_words)
    common_words = word_counts.most_common(20)
    
    words_df = pd.DataFrame(common_words, columns=['word', 'count'])

    plt.figure(figsize=(12, 8))
    sns.barplot(data=words_df, y='word', x='count', palette='magma')
    plt.title('Top 20 Most Frequent Words (excluding stopwords)', fontsize=15)
    plt.xlabel('Frequency', fontsize=12)
    plt.ylabel('Word', fontsize=12)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    visualize_data('./data_phase3/train_cleaned.csv')
    visualize_data('./data_phase3/test_cleaned.csv')    # testsamplesubmission
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import torch
import numpy as np


df_full_train = pd.read_csv('./data_phase3/train_cleaned.csv')

# 
df_submit = pd.read_csv('./data_phase3/submission_cleaned.csv')

# 
label_map = {
    'anger': 0, 'disgust': 1, 'fear': 2, 
    'joy': 3, 'sadness': 4, 'surprise': 5
}
inv_label_map = {v: k for k, v in label_map.items()}

# 
df_full_train = df_full_train.dropna(subset=['emotion', 'text'])
df_full_train['labels'] = df_full_train['emotion'].map(label_map)
df_full_train['labels'] = df_full_train['labels'].astype(int)

# 
# stratify=df_full_train['labels'] 
train_df, eval_df = train_test_split(
    df_full_train, 
    test_size=0.2, 
    random_state=42, 
    stratify=df_full_train['labels']
)

print(f": {len(train_df)}")
print(f": {len(eval_df)} ( F1-score)")

# 
model_args = ClassificationArgs()
model_args.num_train_epochs = 3
model_args.train_batch_size = 16
model_args.eval_batch_size = 32
model_args.learning_rate = 3e-5
model_args.max_seq_length = 128
model_args.overwrite_output_dir = True
model_args.save_model_every_epoch = False
model_args.reprocess_input_data = True
model_args.silent = False 

# 
cuda_available = torch.cuda.is_available()

model = ClassificationModel(
    'bertweet',               #  bertweet
    'vinai/bertweet-base',    # 
    num_labels=6, 
    args=model_args, 
    use_cuda=cuda_available
)

model.train_model(train_df)

# 
result, model_outputs, wrong_predictions = model.eval_model(eval_df)

# metrics
preds = np.argmax(model_outputs, axis=1)
true_labels = eval_df['labels'].values

f1 = f1_score(true_labels, preds, average='macro')
acc = accuracy_score(true_labels, preds)

print(f"Validation Results:")
print(f"Accuracy: {acc:.4f}")
print(f"Macro F1-Score: {f1:.4f}") 

# 
#  list 
to_predict = df_submit['text'].astype(str).tolist()

# 
predictions, raw_outputs = model.predict(to_predict)

#  (0,1,2)  (anger, joy...)
predicted_emotions = [inv_label_map[pred] for pred in predictions]
df_submit['emotion'] = predicted_emotions

#  ( id  emotion)
final_submission = df_submit[['id', 'emotion']]
final_submission.to_csv('./result_phase3/final_submission_bertweet-base.csv', index=False)
```
